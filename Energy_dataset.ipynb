**Description:**

Analyzing the Energy Efficiency of Buildings
The dataset includes features such as surface/wall/roof area, glass area, glass area distribution and orientation of 768 simulated building shapes. Based on this data, it is aimed to estimate the heating and cooling load of the building. **Data Set Information:**

We perform energy analysis using 12 different building shapes. The buildings differ with respect to the glazing area, the glazing area distribution, and the orientation, amongst other parameters. **Attribute Information:**

The dataset contains eight attributes (or features, denoted by X1…X8) and two responses (or outcomes, denoted by Y1 and Y2). The aim is to use the eight features to analyze the dataset and predict each of the two responses.
X1: Relative Compactness, X2: Surface Area, X3: Wall Area, X4: Roof Area, X5: Overall Height, X6: Orientation, X7: Glazing Area, X8: Glazing Area Distribution, Y1: Heating Load, Y2: Cooling Load. A dataset “Energy_dataset.csv” is provided for data analytics.

# Executing SQL queries in R

1. Utilizing Data Manipulation Packages and read.csv:
Use read.csv to import the CSV data from the basic R package.
Use functions from base R or data manipulation packages like dplyr to filter, alter, and summarize the data as needed. These features are comparable to what you could accomplish with the aggregate, group by, WHERE, and SELECT functions in SQL.

2. Using SQL-like Queries with Sqldf:
The sqldf package in R can be used to run SQL queries for a CSV file containing the Energy dataset. Use the sqldf function to write queries that resemble SQL. This enables you to specify operations involving data manipulation using a recognizable SQL syntax.

 * SQL query written in R to select all records from energy dataset where the heating load value is greater than 42.



```r
# SELECT SQL Statement
result <- sqldf("SELECT * FROM energy WHERE Heating_Load > 42")
print(result)
```

* SQL query written in R to calculate total area by using mathematical expressions to add surface area and roof area of building shapes.

```r
# SQL Statement using mathematical expressions
math <- sqldf("SELECT Surface_Area, Roof_Area, Surface_Area + Roof_Area AS Total_Area FROM energy")
print(math)
```

* SQL query written in R to calculate the average heating load for all buildings with a surface area greater than 250 and maximum cooling load value by using Aggregate Functions.

```r
# SQL Statement using aggregate functions
aggr <- sqldf("SELECT AVG(Heating_Load) AS Avg_Heating_Load, MAX(Cooling_Load) AS Max_Cooling_Load FROM energy WHERE Surface_Area > 250")
print(aggr)
```

* SQL query written in R to calculate the square of relative compactness of building shapes whose surface area value is equal to 588 by using Arithmetic Functions.

```r
# SQL Statement using arithmetic functions
arith <- sqldf("SELECT Relative_Compactness, Surface_Area, POWER(Relative_Compactness, 2) AS Compactness_Squared FROM energy WHERE Surface_Area = 588")
print(arith)
```

# Data Visualization

```r
library(ggplot2)  # To create informative and visually appealing plots using the R package for data visualization
```

* Scatterplot of Heating Load vs. Surface Area

```r
ggplot(energy, aes(x = Surface_Area, y = Heating_Load)) +
  geom_point() +
  labs(title = "Heating Load vs. Surface Area",
       x = "Surface Area",
       y = "Heating Load") +
  theme_minimal()
```

* Barplot of Average Heating Load by Building Shape

```r
# Calculate the average heating load by building shape
avg_heating_load <- aggregate(Heating_Load ~ Relative_Compactness, data = energy, FUN = mean)

# Create a barplot of average heating load by building shape
ggplot(avg_heating_load, aes(x = Relative_Compactness, y = Heating_Load)) +
  geom_bar(stat = "identity", fill = "lightblue", color = "black") +
  labs(title = "Average Heating Load by Building Shape",
       x = "Building Shape",
       y = "Average Heating Load") +
  theme_minimal()
```

# Data Manipulation and Transformation in R

❖ R provides an extensive collection of pre-built functions and packages for manipulating and transforming data.

1. Base R Functions

Read data from CSV, text files, etc. via the read.csv() and read.table() functions for data import. Data Filter():Calculate summary statistics (mean, median, etc.) for full datasets or subsets using data aggregation methods like summary() and apply().

2. dplyr Package

Dplyr provides a more readable and succinct method of working with dataframes. It offers verbs for typical tasks:

***filter()***: Choose rows according to criteria.

***arrange()***: Sort rows according to particular columns.

***select()***: Select particular columns.

***mutate()***: Construct new columns by using the existing ones as a basis.

Calculate summary statistics by groups using the ***summarize()*** function (frequently used in conjunction with group_by()).

```r
library(dplyr)
```

* **select()** function

```r
#Selecting columns:
selected_columns <- select(energy, Relative_Compactness, Surface_Area, Heating_Load)
print(selected_columns)
```

* **filter()** function

```r
#Filtering rows:
filtered_data <- filter(energy, Heating_Load > 100)
print(filtered_data)
```

* **arrange()** function

```r
#Sorting rows:
sorted_data <- arrange(energy, desc(Heating_Load))
print(sorted_data)
```

* **mutate()** function

```r
#using rbind() function to insert new rows
new_row <- c(1.2, 300, 200, 100, 5, 90, 50, 0.5, 15, 10)
updated_data <- rbind(energy, new_row)
print(updated_data)
```

```r
#using mutate() function to update rows
updated_data <- mutate(energy, Heating_Load = ifelse(Relative_Compactness==1.2, 20, Heating_Load))
print(updated_data)
```

* **summarize()** and group_by() function

```r
# Calculate the mean heating load for each building shape
energy_data_summary <- energy %>%
  group_by(Relative_Compactness) %>%
  summarise(mean_heating_load = mean(Heating_Load))
print(energy_data_summary)
```
